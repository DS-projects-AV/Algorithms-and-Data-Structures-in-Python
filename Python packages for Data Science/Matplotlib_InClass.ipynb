```python
y1 = x *2 
y2 = x **2

plt.plot(x, y1)
plt.xlabel("small xs")
plt.plot(x, y2)
plt.ylabel("y values for the x")
plt.title("different time/cost growth")
```

```python
y1 = x *2 
y2 = x **2

plt.plot(x, y1)
plt.xlabel("small xs")
plt.plot(x, y2)
plt.ylabel("y values for the x")
plt.title("different time/cost growth")
plt.legend(["Linear", "Square"])
```

```python
y1 = x *2 
y2 = x **2
plt.plot(x, y1, '0')
plt.xlabel("small xs")
plt.plot(x, y2, '*')
plt.ylabel("y values for the x")
plt.title("different time/cost growth")
plt.legend(["Linear", "Square"])
```

```python
y1 = x *2 
y2 = x **2


marks = ['o', '*', '.', ',', 'x', 'X', '+', 'P', 's', 'D', 'd', 'p', 'H', 'h', 'v', '^', '<', '>', '1', '2', '3', '4', '|', '_']

for m in marks:
  plt.plot(x, y1)
  plt.xlabel("small xs")
  print(m)
  plt.plot(x, y2, m)
  
  plt.ylabel("y values for the x")
  plt.title("different time/cost growth")
  plt.legend(["Linear", "Square"])
  plt.show()

```

```python
colors = ['r','g','b','c','m','y','k','w']

y1 = x *2 
y2 = x **2

for c in colors:
  plt.plot(x, y1)
  plt.xlabel("small xs")
  print(c)
  c = '^-.' + c
  plt.plot(x, y2, c)
  
  plt.ylabel("y values for the x")
  plt.title("different time/cost growth")
  plt.legend(["Linear", "Square"])
  plt.show()


```

```python
y1 = x *2 
y2 = x **2
plt.plot(x, y1, 'o')
plt.xlabel("small xs")
plt.plot(x, y2, '*')
plt.ylabel("y values for the x")
plt.title("different time/cost growth")
plt.legend(["Linear", "Square"])
plt.grid()
```

```python
y1 = x *2 
y2 = x **2
plt.subplot(2, 1, 1)
plt.plot(x, y1, 'o')
plt.xlabel("small xs")
plt.ylabel("y values for the x")
plt.title("different time/cost growth")
plt.grid()
plt.subplot(2, 1, 2)
plt.plot(x, y2, '*')
plt.xlabel("small xs")
plt.ylabel("y values for the x")
plt.title("different time/cost growth")
plt.grid()
```

```python
x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99, 86, 87, 111, 86, 103, 87, 94, 78, 77, 85, 86, 99])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
size = np.array([3, 13, 33, 43, 1000, 2, 202, 35, 3, 13, 33, 43,12])

plt.scatter(x, y, c = colors, cmap='hot', s = size,alpha = 0.5 )
plt.colorbar()
```
