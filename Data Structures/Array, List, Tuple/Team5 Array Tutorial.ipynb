## Introduction

1. **Array** 
  
  *Definition:* a data structure that stores values of the same type.

  *Graphical view:*

image.png

An array is a data structure that contains a group of elements. Typically these elements are all of the same data type, such as an integer or string. Arrays are commonly used in computer programs to organize data so that a related set of values can be easily sorted or searched.

For example, a search engine may use an array to store Web pages found in a search performed by the user. When displaying the results, the program will output one element of the array at a time. This may be done for a specified number of values or until all the values stored in the array have been output. While the program could create a new variable for each result found, storing the results in an array is much more efficient way to manage memory.



2. **List** 
  
  *Definition:* a data structure that stores values of different types in a defined order.

3. **Tuple** 
  
  *Definition:* a data structure that stores values of different types and that is immutable.

## Python 3 Running Examples


###Arrays

Arrays are used to store multiple values in one single variable.

```python
import array as arr

#create Array of integer type
a = arr.array("i",[1,2,3,4,5,6])
print(f"\nThe intialised array is : {a}")

#append an element to the array
a.append(7)
print(f"\nThe array after element is appended is : {a}")

#remove a element from a specific index from the array
a.pop(2)

#remove a specific element from the array
a.remove(5)
print(f"\nThe array after elements are removed is : {a}")

#Print length of the array
print(f"\nThe length of the array is : {len(a)}")
```

###Lists

Lists are used to store multiple items in a single variable.

```python
#create a list
a=[10, 20, 'apple']
print(f"\nThe intialised list is : {a}")

#append an element to the list
a.append(40)
print(f"\nThe list after new element appended is : {a}")

#change an element of the list
a[1] = 25
print(f"\nThe list after element changed is : {a}")

#remove an element of the list
a.remove(10)
print(f"\nThe list after element removed is : {a}")

#Print length of the list
print(f"\nThe length of the list is : {len(a)}")
```

###Tuples

A tuple is a collection which is ordered and unchangeable.

```python
a = tuple()

a = (420.42,65,"apple","x")

print(f"\nThe intialised tuple is : {a}")

z=a[2]

print(f"\nThe element with index as 2 in the tuple is : {z}")

# Since the tuples are immutable, we can't add any element and delete any element in a tuple.
```

## Python 3 Implementation 

### Arrays

```python
import array as arr

#Array of integer type
a = arr.array("i",[1,2,3,4,5,6])

#Array of float type
a = arr.array("d",[420.42,69.5,6.69,69.69,42.69])


for i in range(0,5):
    print(a[i])


"""
Array Operations:
1) Accessing 
1) Slicing
2) Adding
3) Removing

"""

# Accessing

print(f"\nAccess the element 3: {a[2]}")

print(f"\nAccess the element 0: {a[0]}")

#Slicing

print(f"\nSlicing the elements 2-4 : {a[2:4]}")

print(f"\nSlicing the elements: {a[:-1]}")

#Adding 

print(f"\nArray before insertion : {a}")

"""
Inserting the element using the insert statement using this
we can insert element anywhere in the array.

We can also use Append to add an element in the array, but
it appends the element at the last position in the array.

"""

a.insert(1,999999.69696969)

print(f"\nArray after insertion : {a}")

#Removing 

print(f"\nArray before removal: {a}")

"""
To remove the elements in an array either we can use the pop
or the remove keywords. a.pop() takes the index of an element to be
removed and a.remove() takes the actual element to be removed
if there are multiple elements that are same in a an array remove 
deletes the first occurence of that element.

"""
a.pop(1)

print(f"\nArray after removal: {a}")



```

### Lists

```python
"""
Below are the two ways to intialize the list

"""

#List of integer type
a = list()
a.append(100)
a.append(69)
a.append(420)

print(f"\nThe intialised list is : {a}")

#List can also have a mix of float, integer and characters in a single list

print (f"\nThe List with float,int, and character vals :{['a',1,4,5,7.8,'d']}")

a = [420.42,69.5,6.69,69.69,42.69]


for i in a:
    print(i)


"""
List Operations:
1) Accessing 
1) Slicing
2) Adding
3) Removing

"""

# Accessing

print(f"\nAccess the element 3: {a[2]}")

print(f"\nAccess the element 0: {a[0]}")

#Slicing

print(f"\nSlicing the elements 2-4 : {a[2:4]}")

print(f"\nSlicing the elements: {a[-1]}")

#Adding 

print(f"\nArray before insertion : {a}")

"""
Inserting the element using the "a.insert(idx,value)" statement using this
we can insert element anywhere in the list.

We can also use "a.append(value)" to add an element in the array, but
it appends the element at the last position in the list.

"""

a.append(999999.69696969)

print(f"\nList after insertion : {a}")

#Removing 

print(f"\nList before removal: {a}")

"""
To remove the elements in an array either we can use the pop
or the remove keywords, "a.pop()"" takes the index of an element to be
removed. 
"a.remove()" takes the actual element to be removed
if there are multiple elements that are same in a an array remove 
deletes the first occurence of that element.

"""
a.remove(999999.69696969)

print(f"\nList after removal: {a}")



```

### Tuples

```python
"""
Below are the two ways to intialize the tuple

"""

#List of integer type
a = tuple()

print(f"\nThe intialised tuple is : {a}")

#Tuple can also have a mix of float, integer and characters in a single list

print (f"\nThe tuple with float,int, and character vals : {('a',1,4,5,7.8,'d')}")

a = (420.42,69.5,6.69,69.69,42.69)

for i in a:
    print(i)


"""
Tuple Operations:
1) Accessing 
2) Slicing

"""

# Accessing

print(f"\nAccess the element 3: {a[2]}")

print(f"\nAccess the element 0: {a[0]}")

#Slicing

print(f"\nSlicing the elements 2-4 : {a[2:4]}")

print(f"\nSlicing the elements: {a[-1]}")

#Adding 

print(f"\nTuple before insertion : {a}")

"""
Since the tuples are immutable we can't edit the elements in a 
tuple i.e we can't add any element and delete any element in a
tuple, the code below will give an error.
"""

a.append(999999.69696969)


```
